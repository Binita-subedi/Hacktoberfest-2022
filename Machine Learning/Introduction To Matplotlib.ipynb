## Now using own DataFrame

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$,\.,]', '')
```

```python
car_sales
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods = len(car_sales))
```

```python
car_sales
```

```python
car_sales['Total Sales'] = car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
car_sales['Total Sales'] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

### Lets plot the total sales

```python
car_sales.plot(x='Sale Date', y='Total Sales');
```

```python
car_sales['Price'] = car_sales["Price"].astype(int)
car_sales.plot(x='Odometer (KM)', y="Price", kind="scatter")
```

# Which one would you use?(Pyplot vs matplotlib OO method)

### *When plotting something quickly, okay to use pyplot method
### *When plotting something more advanced, use OO Method

# Heart-Disease

```python
over_50 = heart_disease[heart_disease['age']>50]
```

```python
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot Method
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

# OO Method

```python
fig, ax = plt.subplots(figsize=(10,6))
```

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
ax.set_xlim([45,100])
```

```python
fig
```

# OO Method

```python
fig, ax = plt.subplots(figsize = (10,6))
```

### Plot the data

```python
scatter = ax.scatter(x=over_50["age"], y=over_50['chol'], c=over_50['target'])
scatter
fig
```

### Customize the plot

```python
ax.set(title = "Heart Disease And Cholestrol Levels", xlabel = "Age", ylabel='Cholestrol');
```

```python
fig
```

### Add a legend

```python
ax.legend(*scatter.legend_elements(), title='target')
```

```python
fig
```

### Add a Horizontal line

```python
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
fig
```
