```python
#Adding Column from series
seats_column = pd.Series([2,4,5,6,4])
```

```python
# Adding column from Python List
fuel_economy = [7.5,9.0,6.0,8.0, 8.5,6,6.5,8,9,8.8]
car_sale['Fuel Per 100 Km'] = fuel_economy
```

```python
car_sale
```

```python
car_sale['Total Fuel Used'] = car_sale['Odometer (KM)']/100 * car_sale["Fuel Per 100 Km"]
```

```python
car_sale
```

```python
#Create a column from a single value
car_sale["Number of wheels"] = 4
car_sale
```
